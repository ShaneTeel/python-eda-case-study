```python
def remove_newline(file_path):
    '''Removes all newlines in a line of text

    Parameters
    ----------
    text: str

    Returns
    -------
    text_no_nl: str

    Examples
    --------
    >>> remove_newline("\nlife happens when youre busy\n making other plans\n")
    'life happens when youre busy making other plans'
    '''
    with open(file_path, 'r') as txt_file:
        text = txt_file.read()
        text_no_nl = text.replace("\n", "")
    return text_no_nl
```

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
stopwords = ENGLISH_STOP_WORDS

def read_the_file(fname):
    f = open(fname, 'r')
    text_string = f.read()
    return text_string

file_path = r"base-python-casestudy/data/train_to_busan_description.txt"
text = read_the_file(file_path)

def split_text_into_words(text):
    text = text.split()
    return text

words = split_text_into_words(text)

def remove_stopwords(word_lst, stopwords_set):
    new_lst = []
    for word in word_lst:
        if word.lower() in stopwords_set:
            continue
        else:
            new_lst.append(word)
    return new_lst

print(remove_stopwords(words, stopwords))
```
